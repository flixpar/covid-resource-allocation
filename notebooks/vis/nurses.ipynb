```python
Fips = namedtuple('Fips', 'div county components')
geography_dir = '../../data/geography'

with open(os.path.join(geography_dir, '0312msa.txt'), 'rb') as f:
    lines = f.readlines()
lines = lines[11:-21]
```

```python
def _validate(entry):
    return entry if len(entry) > 0 else None
```

```python
fips = defaultdict(list)
for line in lines:
    line = line.decode('utf-8',errors='ignore').strip()
    if len(line) == 0: continue
    CBSA = _validate(line[:5].strip())
    line = line[8:]
    DIV = _validate(line[:5].strip())
    line = line[8:]
    County = _validate(line[:5].strip())
    Components = line[8:].strip()
    if County is None:
        key = CBSA
        fips[key] = {'div': [], 'county': [], 'components': []}
        continue
    fips[key]['div'].append(DIV)
    fips[key]['county'].append(County)
    fips[key]['components'].append(Components)
```

```python
cbsa = pd.read_csv(os.path.join(geography_dir, 'cbsas.csv')).astype({'CBSA Code': str})#.loc[:, ['category', 'code', 'title']]
cbsa.columns = ['code', 'title', 'category']
cbsa.head()
```

```python
df = pd.DataFrame(columns=['code', 'county_code', 'div_code', 'county_name'])
for i, (k, v) in enumerate(fips.items()):
    df.loc[i] = [k, v['county'], v['div'], v['components']]
df.head()

cbsa = pd.read_csv(os.path.join(geography_dir, 'cbsas.csv')).astype({'CBSA Code': str})
cbsa.columns = ['code', 'title', 'category']
cbsa = pd.merge(cbsa, df, on='code', how='outer')
cbsa = cbsa[cbsa.category == 'Metropolitan']
cbsa.columns = ['area', 'title', 'category', 'county_code', 'div_code', 'county_name']
cbsa.head()
```

```python
met_area_df = pd.read_excel(os.path.join(datadir, 'oesm19ma/MSA_M2019_dl.xlsx'))
met_area_df = met_area_df[met_area_df.occ_code.isin(occ_codes)] .astype({'area': str})
met_area_df = pd.merge(met_area_df, cbsa, on='area', how='outer')
met_area_df.head()
```

```python
populations = pd.read_excel(os.path.join(datadir, 'co-est2019-annres.xlsx'), skiprows=4, skipfooter=6)
populations.columns = ['Area', 'Census', 'Estimates Base', 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019]
populations['county'] = populations.apply(lambda x: x.Area.split(', ')[0].split(' County')[0][1:], axis=1)
populations['state'] = populations.apply(lambda x: x.Area.split(', ')[1], axis=1)
populations.head()

counties = pd.read_csv(os.path.join(geography_dir, 'county_codes.csv')).astype({'code': str})
counties.code = counties.apply(lambda x: '0' + x.code if len(x.code) < 5 else x.code, axis=1)
counties.columns = ['code', 'county', 'abbrev']
abbrevs = pd.read_csv(os.path.join(geography_dir, 'state_name.csv'))
abbrevs.columns = ['state', 'nickname', 'abbrev']
counties = pd.merge(counties, abbrevs, on='abbrev', how='outer')
counties.head()

populations = pd.merge(populations, counties, on=['county', 'state'], how='inner')
populations.head()
```

```python
df = met_area_df
def get_populations(entry):
    if isinstance(entry, list):
        return [populations[populations.code == str(code)][2019].iat[0] for code in entry if str(code) in populations.code.tolist()]
    return np.nan

df['populations'] = df.apply(lambda x: get_populations(x.county_code), axis=1)
df.head()
```
