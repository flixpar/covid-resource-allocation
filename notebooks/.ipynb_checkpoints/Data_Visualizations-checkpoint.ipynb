```python
FIPS = namedtuple('FIPS', 'PMSA ALT County F City MetroArea')

with open('../data/99mfips.txt', 'r') as f:
    lines = f.readlines()
lines = lines[21:-23]
```

```python
def _validate(entry):
    return entry if len(entry) > 0 else None
```

```python
fips = []
for line in lines:
    line = line.strip()
    if len(line) == 0: continue
    MSA = _validate(line[:4].strip())
    line = line[8:]
    PMSA = _validate(line[:4].strip())
    line = line[8:]
    ALT = _validate(line[:2].strip())
    line = line[8:]
    County = _validate(line[:5].strip())
    line = line[8:]
    F = _validate(line[:1].strip())
    line = line[8:]
    City = _validate(line[:5].strip())
    line = line[5:].strip()
    MetroArea = line
    fips.append(FIPS(MSA, PMSA, ALT, County, F, City, MetroArea))
```
